Se uma pessoa escolhida aleatoriamente for de Alagoas ou do Pará,
qual é a probabilidade de ela ter uma renda superior a 5 mil reais?


```python
import pandas as pd

# Nome do arquivo CSV
nome_arquivo = 'RRLNbvGQQtu5fPdxzFPI_populacao_brasileira.csv.csv'

# Carregar o arquivo CSV
df = pd.read_csv(nome_arquivo)

# Converter a coluna de renda para tipo numérico (caso não esteja já)
df['renda'] = pd.to_numeric(df['renda'], errors='coerce')

# Filtrar pessoas de Alagoas ou do Pará
estados_filtrados = df[df['estado'].isin(['AL', 'PA'])]

# Contar quantas pessoas têm renda superior a 5 mil reais
renda_superior_5000 = estados_filtrados[estados_filtrados['renda'] > 5000]

# Total de pessoas de Alagoas ou do Pará
total_estados_filtrados = len(estados_filtrados)

# Número de pessoas com renda superior a 5 mil reais
num_renda_superior_5000 = len(renda_superior_5000)

# Probabilidade de ter renda superior a 5 mil reais
probabilidade_renda_superior_5000 = num_renda_superior_5000 / total_estados_filtrados

print(f"Probabilidade de ter uma renda superior a 5 mil reais: {probabilidade_renda_superior_5000:.2f}")
```

Descubra a probabilidade de uma pessoa ter ensino superior completo
no estado do Amazonas. Qual a probabilidade da quinta pessoa que
você conversar, que é amazonense, ter ensino superior completo?

```python
import pandas as pd

# Nome do arquivo CSV
nome_arquivo = 'RRLNbvGQQtu5fPdxzFPI_populacao_brasileira.csv.csv'

# Carregar o arquivo CSV
df = pd.read_csv(nome_arquivo)

# Filtrar pessoas do estado do Amazonas
amazonenses = df[df['estado'] == 'AM']

# Contar quantas pessoas têm ensino superior completo
ensino_superior_completo = amazonenses[amazonenses['escolaridade'] == 'Superior']

# Total de pessoas no Amazonas
total_amazonenses = len(amazonenses)

# Número de pessoas com ensino superior completo
num_ensino_superior_completo = len(ensino_superior_completo)

# Probabilidade de ter ensino superior completo no Amazonas
probabilidade_ensino_superior_completo = num_ensino_superior_completo / total_amazonenses

print(f"Probabilidade de uma pessoa ter ensino superior completo no estado do Amazonas: {probabilidade_ensino_superior_completo:.2f}")

# Probabilidade da quinta pessoa que você conversar, sendo amazonense, ter ensino superior completo
# É a mesma probabilidade calculada anteriormente
probabilidade_quinta_pessoa = probabilidade_ensino_superior_completo

print(f"Probabilidade da quinta pessoa amazonense ter ensino superior completo: {probabilidade_quinta_pessoa:.2f}")
```

Considerando a renda das pessoas do nosso conjunto, podemos dizer
que a renda de uma pessoa brasileira está na sua maioria em que faixa
(faça faixa de 1.500 reais)? Qual é a sua função densidade de
probabilidade?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Nome do arquivo CSV
nome_arquivo = 'RRLNbvGQQtu5fPdxzFPI_populacao_brasileira.csv.csv'

# Carregar o arquivo CSV
df = pd.read_csv(nome_arquivo)

# Converter a coluna de renda para tipo numérico (caso não esteja já)
df['renda'] = pd.to_numeric(df['renda'], errors='coerce')

# Definir as faixas de renda
faixas = np.arange(0, df['renda'].max() + 1500, 1500)

# Criar uma coluna com a faixa de renda
df['faixa_renda'] = pd.cut(df['renda'], bins=faixas)

# Contar o número de pessoas em cada faixa de renda
contagem_faixa = df['faixa_renda'].value_counts().sort_index()

# Identificar a faixa com o maior número de pessoas
faixa_mais_frequente = contagem_faixa.idxmax()
quantidade_mais_frequente = contagem_faixa.max()

print(f"Faixa de renda mais frequente: {faixa_mais_frequente}")
print(f"Número de pessoas nessa faixa: {quantidade_mais_frequente}")

# Plotar a função densidade de probabilidade (histograma normalizado)
plt.figure(figsize=(10, 6))
plt.hist(df['renda'].dropna(), bins=faixas, density=True, edgecolor='black')
plt.title('Função Densidade de Probabilidade da Renda')
plt.xlabel('Renda (R$)')
plt.ylabel('Densidade de Probabilidade')
plt.grid(True)
plt.show()
```

Calcule a média e a variância da renda da amostra. Depois faça a
distribuição normal, inclua o gráfico.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Nome do arquivo CSV
nome_arquivo = 'RRLNbvGQQtu5fPdxzFPI_populacao_brasileira.csv.csv'

# Carregar o arquivo CSV
df = pd.read_csv(nome_arquivo)

# Converter a coluna de renda para tipo numérico (caso não esteja já)
df['renda'] = pd.to_numeric(df['renda'], errors='coerce')

# Calcular a média e a variância da renda
media_renda = df['renda'].mean()
variancia_renda = df['renda'].var()

print(f"Média da renda: R$ {media_renda:.2f}")
print(f"Variância da renda: R$ {variancia_renda:.2f}")

# Plotar o histograma da renda
plt.figure(figsize=(10, 6))
plt.hist(df['renda'].dropna(), bins=30, density=True, alpha=0.6, color='g', edgecolor='black', label='Distribuição da Renda')

# Gerar a distribuição normal baseada na média e variância
x = np.linspace(df['renda'].min(), df['renda'].max(), 100)
pdf = norm.pdf(x, media_renda, np.sqrt(variancia_renda))

# Plotar a distribuição normal
plt.plot(x, pdf, 'r-', lw=2, label='Distribuição Normal')

plt.title('Distribuição da Renda e Distribuição Normal')
plt.xlabel('Renda (R$)')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True)
plt.show()
```

Primeiro considere a probabilidade encontrada no nosso conjunto de
pessoas com escolaridade de pós-graduação. Considerando a amostra
de população brasileira com 1 milhão de habitantes, qual a
probabilidade de encontrarmos 243 mil pessoas com pós-graduação?

```python
import pandas as pd

# Nome do arquivo CSV
nome_arquivo = 'RRLNbvGQQtu5fPdxzFPI_populacao_brasileira.csv.csv'

# Carregar o arquivo CSV
df = pd.read_csv(nome_arquivo)

# Calcular a probabilidade de pós-graduação
total_pessoas = len(df)
num_pos_graduacao = len(df[df['escolaridade'] == 'Pós-graduação'])
probabilidade_pos_graduacao = num_pos_graduacao / total_pessoas

print(f"Probabilidade de ter pós-graduação: {probabilidade_pos_graduacao:.4f}")
```

```python
import numpy as np
from scipy.stats import norm

# Número total de pessoas e número desejado
N = 1_000_000
k = 243_000

# Calcular média e variância
mu = N * probabilidade_pos_graduacao
sigma = np.sqrt(N * probabilidade_pos_graduacao * (1 - probabilidade_pos_graduacao))

# Calcular a probabilidade usando a distribuição normal
z_score = (k - mu) / sigma

# Probabilidade de encontrar exatamente k pessoas
probabilidade = norm.cdf(z_score) - norm.cdf(z_score - 1)

print(f"Probabilidade de encontrar exatamente 243.000 pessoas com pós-graduação: {probabilidade:.4f}")

# Para facilitar a visualização
import matplotlib.pyplot as plt

# Plotar a função densidade de probabilidade
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf = norm.pdf(x, mu, sigma)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, 'b-', label='Distribuição Normal')
plt.axvline(k, color='r', linestyle='--', label='k = 243.000')
plt.title('Distribuição Normal da Renda com Pós-Graduação')
plt.xlabel('Número de Pessoas com Pós-Graduação')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True)
plt.show()
```

Somando as densidades nós temos a função de densidade acumulada.
Considerando a coluna ‘Escolaridade’ faça a função de densidade
acumulada discreta para cada nível de escolaridade.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Nome do arquivo CSV
nome_arquivo = 'RRLNbvGQQtu5fPdxzFPI_populacao_brasileira.csv.csv'

# Carregar o arquivo CSV
df = pd.read_csv(nome_arquivo)

# Calcular a frequência relativa de cada nível de escolaridade
frequencia = df['escolaridade'].value_counts(normalize=True).sort_index()

# Calcular a função de densidade acumulada (FDA) discreta
fda = frequencia.cumsum()

# Plotar a FDA para cada nível de escolaridade
plt.figure(figsize=(10, 6))
plt.step(fda.index, fda, where='post', label='Função de Densidade Acumulada')
plt.title('Função de Densidade Acumulada Discreta para Nível de Escolaridade')
plt.xlabel('Nível de Escolaridade')
plt.ylabel('Densidade Acumulada')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

Qual a margem de erro amostral da proporção populacional
considerando a proporção de pessoas com nível de inglês
intermediário?

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

# Nome do arquivo CSV
nome_arquivo = 'RRLNbvGQQtu5fPdxzFPI_populacao_brasileira.csv.csv'

# Carregar o arquivo CSV
df = pd.read_csv(nome_arquivo)

# Calcular a proporção de pessoas com nível de inglês intermediário
total_pessoas = len(df)
num_intermediario = len(df[df['nível de proficiência em inglês'] == 'Intermediário'])
proporcao_intermediario = num_intermediario / total_pessoas

# Definir o tamanho da amostra e o nível de confiança
n = total_pessoas
nivel_confianca = 0.95
z = norm.ppf(1 - (1 - nivel_confianca) / 2)  # Valor crítico para 95% de confiança

# Calcular a margem de erro
margem_erro = z * np.sqrt((proporcao_intermediario * (1 - proporcao_intermediario)) / n)

print(f"Proporção de pessoas com nível de inglês intermediário: {proporcao_intermediario:.4f}")
print(f"Margem de erro amostral: {margem_erro:.4f}")
```

Calcula a renda da população. Qual a probabilidade de encontrar 60
pessoas com uma renda mil reais superior à média?

```python
import pandas as pd
import numpy as np
from scipy.stats import binom

# Nome do arquivo CSV
nome_arquivo = 'RRLNbvGQQtu5fPdxzFPI_populacao_brasileira.csv.csv'

# Carregar o arquivo CSV
df = pd.read_csv(nome_arquivo)

# Calcular a renda média e o desvio padrão
media_renda = df['renda'].mean()
desvio_padrao_renda = df['renda'].std()

# Calcular a renda mil reais acima da média
renda_acima_media = media_renda + 1000

# Proporção de pessoas com renda superior a esse valor
proporcao_acima_media = (df['renda'] > renda_acima_media).mean()

# Definir o tamanho da amostra e a quantidade de pessoas desejadas
tamanho_amostra = len(df)
numero_pessoas = 60

# Calcular a probabilidade usando a distribuição binomial
probabilidade = binom.pmf(numero_pessoas, tamanho_amostra, proporcao_acima_media)

print(f"Renda média: {media_renda:.2f}")
print(f"Desvio padrão da renda: {desvio_padrao_renda:.2f}")
print(f"Renda mil reais acima da média: {renda_acima_media:.2f}")
print(f"Proporção de pessoas com renda superior a esse valor: {proporcao_acima_media:.4f}")
print(f"Probabilidade de encontrar 60 pessoas com renda superior a mil reais acima da média: {probabilidade:.4f}")
```

Qual a probabilidade de escolhermos alguém do Sudeste que seja
homem, com ensino fundamental e com renda maior que 2 mil reais por
mês?

```python
import pandas as pd

# Nome do arquivo CSV
nome_arquivo = 'RRLNbvGQQtu5fPdxzFPI_populacao_brasileira.csv.csv'

# Carregar o arquivo CSV
df = pd.read_csv(nome_arquivo)

# Filtrar as pessoas do Sudeste, que são homens, com ensino fundamental e renda maior que 2 mil reais
filtro = (df['estado'] == 'SP') & \
         (df['sexo'] == 'M') & \
         (df['escolaridade'] == 'Fundamental') & \
         (df['renda'] > 2000)

# Número de pessoas que atendem aos critérios
num_pessoas_criterio = len(df[filtro])

# Total de pessoas no DataFrame
total_pessoas = len(df)

# Calcular a proporção
proporcao = num_pessoas_criterio / total_pessoas

print(f"Número de pessoas do Sudeste, homem, com ensino fundamental e renda maior que 2 mil reais: {num_pessoas_criterio}")
print(f"Total de pessoas: {total_pessoas}")
print(f"Probabilidade de escolher alguém que atende aos critérios: {proporcao:.4f}")
```
